# Init

```python
import datahandler

import matplotlib.pyplot as plt


data_preprocessor = datahandler.DataPreprocessorOUS_V2()
data_preprocessor.execute()

data_loader = datahandler.DataLoader(datahandler.DataPreprocessorOUS_V2)
data_loader.execute(clean=False, processed=True, enhanced=True)
```

# Plot processed data

```python
nan_counts = data_loader.processed_incidents_df.isna().sum()
nan_counts = nan_counts[nan_counts > 0].sort_values(ascending=False)

nan_counts.plot(kind="bar", figsize=(10, 6))
plt.title("Number of NaN Values by Column")
plt.ylabel("Number of NaN Values")
plt.xlabel("Columns")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
condition1 = data_loader.processed_incidents_df["time_ambulance_arrived_at_scene"].notna() & data_loader.processed_incidents_df["time_ambulance_arrived_at_hospital"].isna()
condition2 = data_loader.processed_incidents_df["time_ambulance_arrived_at_scene"].isna() & data_loader.processed_incidents_df["time_ambulance_arrived_at_hospital"].notna()
condition3 = data_loader.processed_incidents_df["time_ambulance_arrived_at_scene"].isna() & data_loader.processed_incidents_df["time_ambulance_arrived_at_hospital"].isna()

num_rows_condition1 = data_loader.processed_incidents_df[condition1].shape[0]
num_rows_condition2 = data_loader.processed_incidents_df[condition2].shape[0]
num_rows_condition3 = data_loader.processed_incidents_df[condition3].shape[0]

# Data
conditions = ["Condition 1", "Condition 2", "Condition 3"]
num_rows = [num_rows_condition1, num_rows_condition2, num_rows_condition3]

# Plot
plt.bar(conditions, num_rows)
plt.ylabel("Incidents")
plt.title("Number of Incidents Affected by Condition")
plt.show()
```

```python
condition1 = data_loader.processed_incidents_df["time_ambulance_dispatch_to_hospital"].notna() & data_loader.processed_incidents_df["time_ambulance_arrived_at_hospital"].isna()
condition2 = data_loader.processed_incidents_df["time_ambulance_dispatch_to_hospital"].isna() & data_loader.processed_incidents_df["time_ambulance_arrived_at_hospital"].notna()
condition3 = data_loader.processed_incidents_df["time_ambulance_dispatch_to_hospital"].isna() & data_loader.processed_incidents_df["time_ambulance_arrived_at_hospital"].isna()
condition4 = data_loader.processed_incidents_df["time_ambulance_arrived_at_scene"].notna() & data_loader.processed_incidents_df["time_ambulance_dispatch_to_hospital"].isna()
condition5 = data_loader.processed_incidents_df["time_ambulance_arrived_at_scene"].notna() & condition3

num_rows_condition1 = data_loader.processed_incidents_df[condition1].shape[0]
num_rows_condition2 = data_loader.processed_incidents_df[condition2].shape[0]
num_rows_condition3 = data_loader.processed_incidents_df[condition3].shape[0]
num_rows_condition4 = data_loader.processed_incidents_df[condition4].shape[0]
num_rows_condition5 = data_loader.processed_incidents_df[condition5].shape[0]

# Data
conditions = ["Condition 1", "Condition 2", "Condition 3", "Condition 4", "Condition 5"]
num_rows = [num_rows_condition1, num_rows_condition2, num_rows_condition3, num_rows_condition4, num_rows_condition5]

# Plot
plt.bar(conditions, num_rows)
plt.ylabel("Incidents")
plt.title("Number of Incidents Affected by Condition")
plt.show()
```
